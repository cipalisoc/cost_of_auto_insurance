```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
filename = '/content/drive/MyDrive/Coding Dojo/Week 4: Explanatory Viz/sheet_1_cost_of_auto_insuranc.csv'
```

```python
# loading dataset
df = pd.read_csv(filename)
df.info()
df.head()
```

```python
# isolating columns into lists
State = df['state']
full = df['full_coverage']
liab = df['minimum_coverage']
```

```python
# barplot showing average cost of auto insurance coverages by state
fig = plt.figure(figsize=(20 , 10))
plt.style.use('seaborn')
plt.xlabel('State')
plt.xticks(rotation=90)
plt.ylabel('Cost ($)')
plt.title('Average Cost of Auto Insurance by State' , fontsize=18)
plt.bar(State, full, color='lightblue' , label='Full Coverage');
plt.bar(State, liab, color='salmon' , label='Liability Only');
plt.legend(loc='upper right');
```
